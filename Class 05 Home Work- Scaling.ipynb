# Min Max Scaler (Normalization) using for loop

```python
#x1.head()
#result_excludingArea=[ i for i in x1 if i!="Area"]
# print(result_excludingArea)
from sklearn.preprocessing import MinMaxScaler
norm_scaler = MinMaxScaler()
for ftr in df_norm_loop :
    if ftr == 'Area':
        continue
    #print(ftr)
    df_norm_loop[[ftr]] = norm_scaler.fit_transform(df_norm_loop[[ftr]]) #fit_transform

df_norm_loop.head()
```

# Standard Scaler (regular & loop)

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
```

```python
# without loop implementation
from sklearn.preprocessing import StandardScaler
stnd_scaler = StandardScaler()

df_stnd[['Marketing Spend']] = stnd_scaler.fit_transform(df_stnd[['Marketing Spend']])
df_stnd[['Administration']] = stnd_scaler.fit_transform(df_stnd[['Administration']])
df_stnd[['Transport']] = stnd_scaler.fit_transform(df_stnd[['Transport']])

df_stnd.head()
```

```python
# with loop implementation
for ftr in df_stnd_loop :
    if ftr == 'Area':
        continue
    #print(ftr)
    df_stnd_loop[[ftr]] = stnd_scaler.fit_transform(df_stnd_loop[[ftr]]) #fit_transform

df_stnd_loop.head()
```
